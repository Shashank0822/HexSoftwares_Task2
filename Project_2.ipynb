# Importing Packages

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

# Loading the Iris Dataset

```python
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data['species'] = iris.target
data['species'] = data['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

# Summary

```python
data.describe()
```

# Checking For Null Values

```python
data.isnull().sum()
```

# Visualization

# Pairplot of Iris Dataset

```python
sns.pairplot(data, hue='species', diag_kind='kde', palette='Set2')
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

# Histogram

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
features = iris.feature_names
for i, ax in enumerate(axes.flat):
    sns.histplot(data[features[i]], ax=ax, color='skyblue')
    ax.set_title(features[i])
plt.tight_layout()
plt.show()
```

# Boxplot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[iris.feature_names], palette='pastel')
plt.title("Boxplot of Features")
plt.xticks(rotation=45)
plt.show()
```

# Correlation Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data[iris.feature_names].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Features")
plt.show()
```

# Violin Plot

```python
plt.figure(figsize=(12, 8))
for feature in features:
    sns.violinplot(x='species', y=feature, data=data, palette='muted')
    plt.title(f"Violin Plot of {feature} by Species")
    plt.show()
```
